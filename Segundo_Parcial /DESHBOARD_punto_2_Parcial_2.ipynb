```python
%%writefile THD.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sym
import control
st.sidebar.title("Presentado por:")
st.sidebar.write("Ever Daniel Hern√°ndez Tulc√°n")


# T√≠tulo y descripci√≥n general
st.title("An√°lisis de un p√©ndulo el√°stico amortiguado")

st.write("""
Las propiedades mec√°nicas del p√©ndulo el√°stico amortiguado se analizan utilizando la teor√≠a de los SLIT. La configuraci√≥n mec√°nica subyacente se muestra en la siguiente figura:


![P√©ndulo El√°stico Amortiguado](https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/damped_spring.png?raw=1)
""")


# Secci√≥n te√≥rica
st.write("""
### Ecuaci√≥n diferencial

La ecuaci√≥n diferencia del sistema mec√°nico se obtiene al considerar el equilibrio de fuerzas ejercidas sobre la masa:
""")

st.latex(r"F_\text{S}(t) + F_\text{F}(t) + F_\text{I}(t) = F_\text{E}(t)")

st.write("""
Donde:

Donde $F_{E}(t) $ denota una fuerza externa que act√∫a sobre la masa. La fuerza $F_{S}(t)$ inducida por el resorte est√° dada por la ley de Hooke
""")
st.latex(r"F_\text{S}(t) = k y(t)")

st.write("""

Es com√∫n modelar la fuerza de fricci√≥n $F_{F}(t)$ inducida por el amortiguador de manera proporcional a la velocidad de la masa
""")
st.latex(r" F_\text{F}(t) = c \frac{d y(t)}{dt}")

st.write("""
La fuerza inercial $F_{I}(t)$ debida a la aceleraci√≥n de la masa se da como
""")
st.latex(r" F_\text{I}(t) = m \frac{d^2 y(t)}{dt^2} ")

st.write("""
Al introducir las fuerzas anteriormente halladas en el equilibrio de fuerzas se obtiene la ecuaci√≥n diferencial que describe el desplazamiento del p√©ndulo el√°stico amortiguado
""")

st.latex(r"m\,\frac{d^{2}y(t)}{dt^{2}} + c\,\frac{dy(t)}{dt} + k\,y(t) = F_{E}(t)")


st.write("""
Como consecuencia de la fuerza externa.

La ecuaci√≥n anterior constituye una EDO con coeficientes constantes. Puede interpretarse como un SLTI con la fuerza externa $x(t)= F_{E}(t)$ como se√±al de entrada y el desplazamiento de la masa como la se√±al de salida $y(t)$.
""")

# Inicializar impresi√≥n simb√≥lica
sym.init_printing()

# Definir s√≠mbolos
t, m, c, k = sym.symbols('t m c k', real=True)
x = sym.Function('x')(t)
y = sym.Function('y')(t)

# Crear la ecuaci√≥n diferencial
ode = sym.Eq(m * y.diff(t, 2) + c * y.diff(t) + k * y, x)

# Mostrar en pantalla
st.write("### Ecuaci√≥n diferencial generada con SymPy:")
st.latex(sym.latex(ode))

st.write("""
Tomando transformada de Laplace (condiciones iniciales cero) y usando
$Y(s)$ Y $F(s)$ donde: """)


st.latex(r" H(s) = \frac{Y(s)}{F(s)}")

st.write("Funci√≥n de transferencia")

st.latex(r"H(s)=\frac{1}{m s^2 + c s + k}")


# T√≠tulo y descripci√≥n general
st.title("An√°lisis de un p√©ndulo el√°stico amortiguado")

st.write("""
Las propiedades mec√°nicas del p√©ndulo el√°stico amortiguado se analizan utilizando la teor√≠a de los SLIT. La configuraci√≥n mec√°nica subyacente se muestra en la siguiente figura:


![P√©ndulo El√°stico Amortiguado](https://github.com/amalvarezme/SenalesSistemas/blob/master/Otros/circuitoRLC.jpeg?raw=2)
""")
st.write("Calculamos la Ecuaci√≥n Diferencial")

st.write("La ecuaci√≥n diferencial que describe la relaci√≥n entrada/salida en el circuito se obtiene al aplicar las leyes de Kirchhoff. As√≠, se obtiene la siguiente EDO")

st.latex(r"il(t) = ic(t) + iR(t)")


st.latex(r"CL \frac{d^2 Vo(t)}{dt^2} + \frac{L}{R}\frac{d Vo(t)}{dt} + Vo(t) = Vi(t)")

st.write("Aplicando la trasfomada de laplace tenemos")

st.latex(r"CLs^2Vo(s)+ (LsVo(s)/R) + Vo(s) = Vi(s)")

st.write("""
Tomando transformada de Laplace (condiciones iniciales cero) y usando
$Y(s)$ Y $F(s)$ donde: """)


st.latex(r" H(s) = \frac{Y(s)}{F(s)}")

st.write("Funci√≥n de transferencia")

st.latex(r"H(s)=\frac{1}{LC s^2 + \frac{Ls}{R} + 1}")

st.write("Comparaci√≥n con la red el√©ctrica")

st.write("""
Comparando las EDO del p√©ndulo el√°stico amortiguado y el circuito RLC, se tiene que ambos constituyen una EDO de segundo orden con coeficientes constantes.

La comparaci√≥n de la EDO anterior con la del p√©ndulo el√°stico amortiguado produce equivalencia entre sistemas

|                         | Circuito RLC | P√©ndulo el√°stico                          |
|-------------------------|------------------------------------|-------------------------------------------|
| Se√±al de entrada $x(t)$ | $V_{i}(t) = F_{E}(t) C$  |$F_{E}(t) = 1/C*V_{i}(t)$ |
| Se√±al de salida $y(t)$  | $V_{o}(t)$                    | $y(t)$                                    |
|                         | $L=m$                              | $m=L$                                     |
|                         | $R=L/Cc$                              | $c=L/RC$                                     |
|                         | $C= 1/k$                    | $k={1}/{C}$                           |
""")

#___________________________________________________________


# ---------------------------
#  Secci√≥n a√±adida: integraci√≥n del contenido de 2.py
#  (Interfaz interactiva y simulaciones)
# ---------------------------

#  IMPORTACI√ìN DE LIBRER√çAS ADICIONALES (si no est√°n ya importadas)
# (las principales ya est√°n importadas arriba: streamlit, numpy, matplotlib, scipy, sympy)
import scipy.signal as signal  # alias consistente
import math

#  CONFIGURACI√ìN DE LA P√ÅGINA (ajuste no intrusivo)
st.set_page_config(page_title="Simulaci√≥n RLC 2¬∫ Orden", layout="wide", initial_sidebar_state="expanded")

# Sidebar (a√±adimos info adicional sin quitar la del autor)
st.sidebar.write("---")
st.sidebar.title("Simulaci√≥n adicional")
st.sidebar.write("Basado en: 2.py ‚Äî integraci√≥n en THD.py")

#titulo de la nueva secci√≥n
st.header("Simulaci√≥n sistema masa‚Äìresorte‚Äìamortiguador y RLC paralelo (Interactivo)")

st.write("""
Esta herramienta permite simular un sistema masa‚Äìresorte‚Äìamortiguador y su equivalente
el√©ctrico RLC en paralelo.
""")

st.markdown(r"""
**Aqu√≠ se calculan:**

- Tipo de amortiguamiento: subamortiguado, cr√≠tico y sobreamortiguado.
- Frecuencia natural no amortiguada $\omega_n$ y amortiguada $\omega_d$.
- Factor de amortiguamiento $\zeta$.
- Tiempos caracter√≠sticos: pico $t_p$, levantamiento $t_r$ y establecimiento $t_s$.
""")

# TEOR√çA (expander)
with st.expander("Modelo mec√°nico y equivalente RLC paralelo (Resumen)"):
    st.markdown("### Modelo masa‚Äìresorte‚Äìamortiguador")
    st.latex(r"m\ddot{y}(t) + c\dot{y}(t) + k y(t) = F_E(t)")
    st.markdown("""
    - Entrada mec√°nica: \\(x(t) = F_E(t)\\) (fuerza externa).
    - Salida: \\(y(t)\\) (desplazamiento de la masa).
    """)
    st.markdown("**Funci√≥n de transferencia mec√°nica (condiciones iniciales cero):**")
    st.latex(r"H_m(s) = \frac{Y(s)}{F_E(s)} = \frac{1}{m s^2 + c s + k}")

    st.markdown("### Equivalente RLC en paralelo")
    st.latex(r"H_{RLC}(s) = \frac{V_o(s)}{V_i(s)} = \frac{1}{LC s^2 + \frac{L}{R}s + 1}")
    st.markdown("Comparando denominadores, se obtiene la equivalencia:")
    st.latex(r"m = LC,\quad c = \frac{L}{R},\quad k = 1")
    st.markdown("""
    En este panel se muestran los valores equivalentes \\(R, L, C\\) calculados
    a partir de \\(m\\) y \\(c\\) usando estas relaciones.
    """)

# FUNCI√ìN PARA M√âTRICAS DE ESCAL√ìN
def calcular_metricas_escalon(t_arr, y_arr, valor_final=1.0):
    """
    Calcula:
    - Tiempo de levantamiento (Tr) ~ de 10% a 90%.
    - M√°ximo sobreimpulso (Mp) [%].
    - Tiempo pico (Tp).
    - Tiempo de establecimiento (Ts) al 2 % (√∫ltima vez que sale de la banda).
    """
    y = np.array(y_arr)
    t = np.array(t_arr)

    # Tiempo de levantamiento (10% a 90%)
    try:
        idx_10 = np.where(y >= 0.1 * valor_final)[0][0]
        idx_90 = np.where(y >= 0.9 * valor_final)[0][0]
        t_r = t[idx_90] - t[idx_10]
    except:
        t_r = np.nan

    # M√°ximo sobreimpulso y tiempo pico
    val_max = np.max(y)
    if val_max > valor_final + 1e-12:
        M_p = (val_max - valor_final) / valor_final * 100.0
        idx_pico = np.argmax(y)
        t_p = t[idx_pico]
    else:
        M_p = 0.0
        t_p = np.nan

    # Tiempo de establecimiento (2%): buscamos la √∫ltima vez que sale de banda ¬±2%
    banda = 0.02 * valor_final
    t_s = np.nan
    for i in range(len(y)-1, -1, -1):
        if abs(y[i] - valor_final) > banda:
            t_s = t[i]
            break

    return t_r, M_p, t_p, t_s

# SECCI√ìN: CONTROLES EN SIDEBAR (interactivo)
st.sidebar.header("Par√°metros del sistema (simulaci√≥n)")

tipo_sistema = st.sidebar.selectbox(
    "Tipo de sistema",
    ["Subamortiguado (0 < Œ∂ < 1)", "Amortiguamiento cr√≠tico (Œ∂ = 1)", "Sobreamortiguado (Œ∂ > 1)"]
)

modo_lazo = st.sidebar.radio(
    "Tipo de lazo",
    ["Lazo abierto", "Lazo cerrado"],
    help="En lazo cerrado se usa realimentaci√≥n unitaria: T(s) = H(s)/(1 + H(s))."
)

# Par√°metros mec√°nicos b√°sicos con sliders (no interfieren con el bloque original)
m_val = st.sidebar.slider("Masa m", 0.1, 10.0, 1.0, 0.1, key="m_sim")
k_val = st.sidebar.slider("Constante del resorte k", 0.1, 20.0, 1.0, 0.1, key="k_sim")

# Selecci√≥n / rango del factor de amortiguamiento Œ∂
if tipo_sistema.startswith("Subamortiguado"):
    zeta = st.sidebar.slider("Factor de amortiguamiento Œ∂ (0 < Œ∂ < 1)", 0.05, 0.95, 0.3, 0.05, key="zeta_sub")
elif tipo_sistema.startswith("Sobreamortiguado"):
    zeta = st.sidebar.slider("Factor de amortiguamiento Œ∂ (Œ∂ > 1)", 1.05, 5.0, 2.0, 0.05, key="zeta_sobre")
else:  # Cr√≠tico
    zeta = 1.0
    st.sidebar.info("Para amortiguamiento cr√≠tico se fija Œ∂ = 1")

# SECCI√ìN: C√ÅLCULOS DEL MODELO (num√©ricos)
omega_n = np.sqrt(k_val / m_val)
c_val = 2 * zeta * np.sqrt(k_val * m_val)

if 0 < zeta < 1:
    omega_d = omega_n * np.sqrt(1 - zeta**2)
else:
    omega_d = np.nan

if 0 < zeta < 1:
    t_p_analitico = np.pi / omega_d
    t_s_analitico = 3 / (zeta * omega_n)
else:
    t_p_analitico = np.nan
    t_s_analitico = 3 / (zeta * omega_n)

# Equivalente RLC paralelo (usamos L fijo = 1 H para visualizar)
L_eq = 1.0
C_eq = m_val / L_eq if m_val > 0 else np.nan
R_eq = L_eq / c_val if c_val > 0 else np.nan

# Mostrar par√°metros como m√©tricas
st.subheader("Par√°metros del modelo (simulaci√≥n)")
col_mec1, col_mec2, col_mec3 = st.columns(3)
col_mec1.metric("m (masa)", f"{m_val:.3g}")
col_mec2.metric("k (resorte)", f"{k_val:.3g}")
col_mec3.metric("c (amortiguador)", f"{c_val:.3g}")

col_dyn1, col_dyn2, col_dyn3 = st.columns(3)
col_dyn1.metric("œâ‚Çô (natural)", f"{omega_n:.4g} rad/s")
col_dyn2.metric("Œ∂", f"{zeta:.4g}")
col_dyn3.metric("œâ_d (amortig.)", f"{omega_d:.4g} rad/s" if 0 < zeta < 1 else "No aplica")

col_tiempo1, col_tiempo2 = st.columns(2)
col_tiempo1.metric("t‚Çö (anal√≠tico)", f"{t_p_analitico:.4f} s" if not np.isnan(t_p_analitico) else "No aplica")
col_tiempo2.metric("t‚Çõ (anal√≠tico)", f"{t_s_analitico:.4f} s")

st.markdown("**Equivalente RLC paralelo (m = L¬∑C, c = L/R, k = 1):**")
col_rlc1, col_rlc2, col_rlc3 = st.columns(3)
col_rlc1.metric("R (Œ©)", f"{R_eq:.4g}" if not np.isnan(R_eq) else "No definido")
col_rlc2.metric("L (H)", f"{L_eq:.4g}")
col_rlc3.metric("C (F)", f"{C_eq:.4g}" if not np.isnan(C_eq) else "No definido")

# Texto tipo y lazo
texto_tipo = ""
if 0 < zeta < 1:
    texto_tipo = "Sistema subamortiguado: respuesta r√°pida con oscilaciones y posible sobreimpulso."
elif np.isclose(zeta, 1.0):
    texto_tipo = "Sistema cr√≠ticamente amortiguado: respuesta m√°s r√°pida posible sin oscilaciones."
elif zeta > 1:
    texto_tipo = "Sistema sobreamortiguado: respuesta mon√≥tona, muy estable pero m√°s lenta."
else:
    texto_tipo = "Valor de Œ∂ fuera del rango t√≠pico de an√°lisis cl√°sico."

texto_lazo = "Lazo abierto: muestra la din√°mica natural de la planta."
if modo_lazo == "Lazo cerrado":
    texto_lazo = "Lazo cerrado: la realimentaci√≥n corrige la salida y suele reducir el error y el sobreimpulso."

st.info(f"{texto_tipo}\n\n{texto_lazo}")

# SECCI√ìN: FUNCI√ìN DE TRANSFERENCIA (NUM√âRICA PARA SIMULACI√ìN)
# Forma can√≥nica: H(s) = œân^2 / (s^2 + 2Œ∂œân s + œân^2)
num_num = [omega_n**2]
if modo_lazo == "Lazo abierto":
    den_num = [1.0, 2.0*zeta*omega_n, omega_n**2]
else:
    # Para lazo cerrado T(s) = H/(1+H); construimos num y den simb√≥licamente y luego polinomios num√©ricos
    # Simplificaci√≥n pr√°ctica: formaremos lazo cerrado num/den mediante algebra simb√≥lica segura
    s_sym = sym.Symbol('s')
    H_sym = (omega_n**2) / (s_sym**2 + 2*zeta*omega_n*s_sym + omega_n**2)
    T_sym = sym.simplify(H_sym / (1 + H_sym))
    # Obtener numerador y denominador simb√≥licos
    num_c_sym, den_c_sym = sym.fraction(sym.simplify(T_sym))
    # Convertimos a listas num√©ricas
    num_c_poly = sym.Poly(num_c_sym, s_sym)
    den_c_poly = sym.Poly(den_c_sym, s_sym)
    num_num = [float(coef) for coef in num_c_poly.all_coeffs()]
    den_num = [float(coef) for coef in den_c_poly.all_coeffs()]

# Si estamos en lazo abierto y num_num sigue siendo lista simple, aseguramos floats:
if isinstance(num_num, list):
    num_num = [float(x) for x in num_num]
if isinstance(den_num, list):
    den_num = [float(x) for x in den_num]

# Construcci√≥n del sistema num√©rico con scipy.signal
sistema_num = signal.TransferFunction(num_num, den_num)

# L√≠mite de tiempo para las gr√°ficas
if omega_n > 0 and abs(zeta) > 0.05:
    t_lim = max(5.0, 15.0 / (omega_n * abs(zeta)))
else:
    t_lim = 10.0

st.markdown("---")
st.success("Gr√°ficas generadas con SciPy para el sistema seleccionado")

# SECCI√ìN: GR√ÅFICAS (BODE, polos/ceros, TIEMPO)

# 1. Diagrama de Bode (usando scipy.signal.bode)
st.subheader("1. Diagrama de Bode")
w, mag, phase = signal.bode(sistema_num)
fig_bode, (axm, axp) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
axm.semilogx(w, mag)
axm.set_ylabel("Magnitud (dB)")
axm.grid(True, which="both", ls=":", alpha=0.5)
axp.semilogx(w, phase)
axp.set_ylabel("Fase (grados)")
axp.set_xlabel("Frecuencia (rad/s)")
axp.grid(True, which="both", ls=":", alpha=0.5)
st.pyplot(fig_bode)
plt.clf()

# 2. Mapa de polos y ceros (usando ra√≠ces de polinomios)
st.subheader("2. Mapa de Polos y Ceros")
pz_fig, pz_ax = plt.subplots(figsize=(6,4))
# Polos y ceros
try:
    z_roots = np.roots(num_num) if len(num_num) > 1 else np.array([])
    p_roots = np.roots(den_num)
except Exception:
    z_roots = np.array([])
    p_roots = np.array([])

if z_roots.size:
    pz_ax.scatter(np.real(z_roots), np.imag(z_roots), marker='o', label='Ceros')
if p_roots.size:
    pz_ax.scatter(np.real(p_roots), np.imag(p_roots), marker='x', label='Polos')
pz_ax.axhline(0, color='k', linewidth=0.5)
pz_ax.axvline(0, color='k', linewidth=0.5)
pz_ax.set_xlabel("Parte real")
pz_ax.set_ylabel("Parte imaginaria")
pz_ax.set_title("Mapa polos y ceros")
pz_ax.legend()
pz_ax.grid(True, alpha=0.3)
st.pyplot(pz_fig)
plt.clf()

# 3. Respuestas en el tiempo (impulso, escal√≥n, rampa)
st.subheader("3. Respuestas en el dominio del tiempo")

# Tiempo de simulaci√≥n
t_sim = np.linspace(0, t_lim, 2000)

# ----------------------------------------
# Respuesta al impulso
# ----------------------------------------
st.markdown("### ‚û§ Respuesta al impulso")
try:
    t_imp, y_imp = signal.impulse(sistema_num, T=t_sim)
    fig_imp, ax_imp = plt.subplots()
    ax_imp.plot(t_imp, y_imp)
    ax_imp.set_xlabel("Tiempo [s]")
    ax_imp.set_ylabel("Respuesta impulso")
    ax_imp.grid(True, alpha=0.3)
    st.pyplot(fig_imp)
    plt.clf()
except Exception as e:
    st.write("No se pudo calcular la respuesta al impulso:", e)

# ----------------------------------------
# Respuesta al escal√≥n
# ----------------------------------------
st.markdown("### ‚û§ Respuesta al escal√≥n")
try:
    t_step, y_step = signal.step(sistema_num, T=t_sim)
    fig_step, ax_step = plt.subplots()
    ax_step.plot(t_step, y_step)
    ax_step.set_xlabel("Tiempo [s]")
    ax_step.set_ylabel("Respuesta escal√≥n")
    ax_step.grid(True, alpha=0.3)
    st.pyplot(fig_step)
    plt.clf()
except Exception as e:
    st.write("No se pudo calcular la respuesta al escal√≥n:", e)

# ----------------------------------------
# Respuesta a la rampa
# ----------------------------------------
st.markdown("### ‚û§ Respuesta a la rampa")
try:
    # r(t) = t (rampa): convoluci√≥n con respuesta al impulso
    t_imp_r, h_imp = signal.impulse(sistema_num, T=t_sim)
    r = t_sim
    y_ramp = np.convolve(h_imp, r)[:len(t_sim)] * (t_sim[1] - t_sim[0])

    fig_ramp, ax_ramp = plt.subplots()
    ax_ramp.plot(t_sim, y_ramp)
    ax_ramp.set_xlabel("Tiempo [s]")
    ax_ramp.set_ylabel("Respuesta rampa")
    ax_ramp.grid(True, alpha=0.3)
    st.pyplot(fig_ramp)
    plt.clf()
except Exception as e:
    st.write("No se pudo calcular la respuesta a la rampa:", e)


# SECCI√ìN: M√âTRICAS A PARTIR DEL ESCAL√ìN (NUM√âRICO)
st.subheader("4. M√©tricas de desempe√±o (a partir de la respuesta al escal√≥n)")

# Si no calculamos y_step arriba por alg√∫n error, intentar de nuevo
try:
    _ = y_step
except NameError:
    t_step, y_step = signal.step(sistema_num, T=t_sim)

tr_calc, mp_calc, tp_calc, ts_calc = calcular_metricas_escalon(t_step, y_step)

col_met1, col_met2 = st.columns(2)

txt_tr = f"{tr_calc:.4f} s" if not np.isnan(tr_calc) else "No definido"
txt_mp = f"{mp_calc:.2f} %"
txt_tp = f"{tp_calc:.4f} s" if not np.isnan(tp_calc) else "No aplica"
txt_ts = f"{ts_calc:.4f} s" if not np.isnan(ts_calc) else "No converge"

col_met1.metric("Tiempo de levantamiento Tr (10‚Äì90%)", txt_tr)
col_met1.metric("M√°ximo sobreimpulso Mp", txt_mp)
col_met2.metric("Tiempo pico Tp", txt_tp)
col_met2.metric("Tiempo de establecimiento Ts (2%)", txt_ts)

### üìå Conclusi√≥n general sobre las tres respuestas del sistema
st.header("""
Las tres gr√°ficas obtenidas permiten comprender c√≥mo se comporta el sistema din√°mico frente a diferentes tipos de se√±ales:

1. **Respuesta al impulso**
   Esta gr√°fica representa la din√°mica m√°s fundamental del sistema, ya que muestra c√≥mo reacciona cuando recibe una entrada instant√°nea y muy breve.
   Revela si el sistema es oscilatorio, sobreamortiguado o cr√≠ticamente amortiguado, y permite identificar la rapidez con la que disipa energ√≠a.

2. **Respuesta al escal√≥n**
   Modela la reacci√≥n del sistema cuando se aplica una entrada constante desde t = 0.
   Es la m√°s √∫til para evaluar estabilidad, tiempo de establecimiento y si existe sobreimpulso.
   En un sistema sobreamortiguado, la respuesta es lenta y mon√≥tona, sin oscilaciones.

3. **Respuesta a la rampa**
   Representa c√≥mo el sistema sigue una entrada creciente linealmente con el tiempo.
   Esta gr√°fica evidencia la capacidad del sistema para "seguir" se√±ales que cambian continuamente, mostrando el error din√°mico y la limitaci√≥n natural del sistema al responder a variaciones sostenidas.

En conjunto, estas tres respuestas permiten evaluar la din√°mica completa del sistema: su rapidez, estabilidad, capacidad de seguimiento y comportamiento frente a distintos tipos de se√±ales que aparecen en aplicaciones reales.
""")

# GR√ÅFICA COMPARATIVA (sub/crit/sobre)
st.markdown("## GR√ÅFICA COMPARATIVA")
st.markdown(r"""
Esta gr√°fica muestra c√≥mo reacciona el sistema ante una entrada escal√≥n variando el factor de amortiguamiento ($\zeta$):
""")

st.subheader("Comparaci√≥n visual de los tipos de amortiguamiento")

# Par√°metros base
m_base = 1.0
k_base = 1.0
wn_base = np.sqrt(k_base / m_base)

# Tres valores de c que representan los casos
c_sub   = 0.6
c_crit  = 2.0
c_sobre = 4.0

z_sub   = c_sub   / (2 * np.sqrt(k_base * m_base))
z_crit  = c_crit  / (2 * np.sqrt(k_base * m_base))
z_sobre = c_sobre / (2 * np.sqrt(k_base * m_base))

num = [wn_base**2]
den_sub   = [1, 2 * z_sub   * wn_base, wn_base**2]
den_crit  = [1, 2 * z_crit  * wn_base, wn_base**2]
den_sobre = [1, 2 * z_sobre * wn_base, wn_base**2]

sys_sub   = signal.TransferFunction(num, den_sub)
sys_crit  = signal.TransferFunction(num, den_crit)
sys_sobre = signal.TransferFunction(num, den_sobre)

t_cmp = np.linspace(0, 12, 2000)
t_sub,   y_sub   = signal.step(sys_sub,   T=t_cmp)
t_crit,  y_crit  = signal.step(sys_crit,  T=t_cmp)
t_sobre, y_sobre = signal.step(sys_sobre, T=t_cmp)

fig_cmp, ax_cmp = plt.subplots(figsize=(7, 4))
ax_cmp.plot(t_sub,   y_sub,   label=fr"Subamortiguado  ($\zeta \approx {z_sub:.2f}$)")
ax_cmp.plot(t_crit,  y_crit,  label=fr"Cr√≠tico         ($\zeta = {z_crit:.1f}$)")
ax_cmp.plot(t_sobre, y_sobre, label=fr"Sobreamortiguado ($\zeta \approx {z_sobre:.1f}$)")

ax_cmp.axhline(1.0, color="gray", linestyle="--", linewidth=1)
ax_cmp.set_xlabel("Tiempo [s]")
ax_cmp.set_ylabel("Respuesta al escal√≥n")
ax_cmp.set_title("Comparaci√≥n de respuestas para distintos factores de amortiguamiento")
ax_cmp.grid(True, alpha=0.3)
ax_cmp.legend(loc="best")

st.pyplot(fig_cmp)
plt.clf()


### üìå Conclusi√≥n sobre la gr√°fica comparativa de amortiguamiento
st.header("""
La gr√°fica permite visualizar c√≥mo cambia la respuesta de un sistema de segundo orden cuando var√≠a el factor de amortiguamiento (\(\zeta\)):

1. **Subamortiguado (\(\zeta < 1\))**
   El sistema presenta oscilaciones debido a que la energ√≠a almacenada en el resorte no se disipa suficientemente r√°pido.
   La respuesta supera temporalmente el valor final (sobreimpulso) y tarda m√°s tiempo en estabilizarse.

2. **Amortiguamiento cr√≠tico (\(\zeta = 1\))**
   Es el punto exacto donde el sistema deja de oscilar pero alcanza el valor final en el menor tiempo posible.
   Representa el comportamiento √≥ptimo cuando se busca rapidez sin sobreimpulso.

3. **Sobreamortiguado (\(\zeta > 1\))**
   El sistema no oscila, pero su respuesta es lenta.
   La disipaci√≥n de energ√≠a es excesiva, causando que el sistema tarde m√°s en llegar al estado estable.

---

### ‚úîÔ∏è Interpretaci√≥n general

La comparaci√≥n evidencia que **el amortiguamiento controla la velocidad y la estabilidad del sistema**:

- Si hay **demasiado poco amortiguamiento**, el sistema responde r√°pido pero oscila.
- Si hay **amortiguamiento cr√≠tico**, el sistema es r√°pido y estable.
- Si hay **demasiado amortiguamiento**, el sistema es estable pero lento.

En resumen, la gr√°fica muestra c√≥mo el valor de \(\zeta\) determina si el sistema prioriza **rapidez**, **estabilidad** o **suavidad** en la respuesta.
""")
```
